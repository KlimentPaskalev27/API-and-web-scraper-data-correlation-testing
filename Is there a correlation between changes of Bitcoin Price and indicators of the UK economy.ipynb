### 3.2 Data – Web Scrape
#### 3.2.1 Source - Macrotrends.net
The data for the UK macroeconomic indices will be retrieved from the macrotrends.net website by implementing a web scrapper. The website offers freely available data on each country’s factors of economic development. Historical data is available for each year, formatted and presented in tables. The web scrapper will iterate over several pages and target the appropriate table with data and return a list of its contents. 
#### 3.2.2 Data Types
The main data for Bitcoin will be the closing price of one Bitcoin (BTC) at the end of a particular date.
The macroeconomic data for the UK will be:

- Manufacturing Output in Billions USD

#### 3.2.3 Retrieving data
Beautiful soup is used to process the received lxml data and convert it into an usable data type. 
It creates a beautiful soup object and passes it the page’s structure as data. 
To receive the data in the first place, the following function is defined in order to identify which elements of the html document are going to be kept or discarded:

```python
# define function to take lxml data from scrapped page and create a data frame with it
def tableToDataFrame(table):
        col_names = ["index","1","2"]   
        row = table.find("tr")
        columns = row.find_all(["td", "th"])  
        for column in columns:    
            text = column.find(text = True)  
            text = text.strip()
            col_names.append(text)    
        df = pd.DataFrame(columns = col_names)   
        
        row_index = 0
        first_row = True

        for row in table.find_all("tr"):
            if first_row:
                first_row = False 
                continue

            df.loc[row_index] = ""
            columns = row.find_all(["td", "th"])

            column_index = 1
            for column in columns:
                text = "".join(column.findAll(text=True, recursive="True")).strip()
                df.iat[row_index, column_index] = text
                column_index += 1
            row_index += 1
        return df
```

#### 3.2.4 Data processing

Followed by a definition of the scrapper() function which will use the http, urllib3 and beautifulsoup libraries to scrape any URLs given as input. It will then use the tableToDataFrame() function defined above to convert the soup object into a Dataframe which can be used to read the data easily.

```python
# function to take input url and scrape the table with contents
def scrapper(url):
    #pull the data from the webpage
    http = urllib3.PoolManager()
    response = http.request("GET", url)

    #create a beautiful soup object, pass in page data
    soup = BeautifulSoup(response.data, "lxml")

    #get the table items. There are 2 tables with same classes on the pages so we neeed the second one always. add [1]
    table = soup.findAll('table', { "class" : "historical_data_table table table-striped table-bordered"})[1]
    
    if  "gdp-per-capita" in url:
        f = open("gdp-per-capita_Data_raw.txt", "w")
        f.write(str(table))
        f.close()
    elif "manufacturing-output" in url:
        f = open("manufacturing-output_Data_raw.txt", "w")
        f.write(str(table))
        f.close()
    elif "inflation-rate-cpi" in url:
        f = open("inflation-rate-cpi_Data_raw.txt", "w")
        f.write(str(table))
        f.close()
    elif "gni-per-capita" in url:
        f = open("gni-per-capita_Data_raw.txt", "w")
        f.write(str(table))
        f.close()
        
    
    
    #use above defined function to convert data into a DataFrame and return it
    return tableToDataFrame(table)
```

#### 3.2.5 Data cleansing and formatting
To make sure the data does not include fields unsuitable for further analysis, i.e. empty or Null, these data cleansing methods are used. <br>
In the created DataFrame, the first row contains what is desired to be the columns title. Therefore, we omit the first row. Then, the DataFrame is filtered by values in the cells inside the Year column. As per this study’s timeframe of analysis, the years between 2009 and 2020 (excl.) are only needed. Finally, the rows are sorted by values in Year ascending.<br>
In order to make the data available to plot using Pandas, the dollar sign ($), coma (,) and percent (%) signs are removed from the strings in each cell in the appropriate columns. The values for GDP are then converted into an integer type, whereas the growth rate is converted into a float type. <br>
In addition, depending which page has been scraped and the column names of the scraped data, the formatter() function will rename the columns to achieve better readibility.

```python
# define a sanitizer and formatting function based on each case url
def formatter(df, url):
    if  "gdp-per-capita" in url:
        
        df = df.iloc[1: , :]
        df = df.rename(columns={"index":"","1": "Year", "2": "GDP Per Capita USD", "U.K. GDP Per Capita - Historical Data": "Annual Growth Rate"})
        df['GDP Per Capita USD'] = df['GDP Per Capita USD'].str.replace('$','')
        df['GDP Per Capita USD'] = df['GDP Per Capita USD'].str.replace(',','').astype(int)
        df['Annual Growth Rate'] = df['Annual Growth Rate'].str.replace('%','').astype(float)
    
    elif "manufacturing-output" in url:
        
        df = df.iloc[1: , :]
        df = df.rename(columns={"index":"","1": "Year", "2": "Manufacturing Output in Billions USD", "U.K. Manufacturing Output - Historical Data": "Manufacturing Output as % of GDP"})
        df['Manufacturing Output in Billions USD'] = df['Manufacturing Output in Billions USD'].str.replace('$','')
        df['Manufacturing Output in Billions USD'] = df['Manufacturing Output in Billions USD'].str[:3].astype(int)
        df['Manufacturing Output as % of GDP'] = df['Manufacturing Output as % of GDP'].str.replace('%','').astype(float)
    
    elif "inflation-rate-cpi" in url:
        
        df = df.iloc[1: , :]
        df = df.rename(columns={"index":"","1": "Year", "2": "Inflation Rate %", "U.K. Inflation Rate - Historical Data": "Annual Change %"})
        df['Inflation Rate %'] = df['Inflation Rate %'].str.replace('%','').astype(float)
        df['Annual Change %'] = df['Annual Change %'].str.replace('%','').astype(float)
        
    elif "gni-per-capita" in url:
        
        df = df.iloc[1: , :]
        df = df.rename(columns={"index":"","1": "Year", "2": "GNI Per Capita USD", "U.K. GNI Per Capita - Historical Data": "Annual % Growth"})
        df['GNI Per Capita USD'] = df['GNI Per Capita USD'].str.replace('$','')
        df['GNI Per Capita USD'] = df['GNI Per Capita USD'].str.replace(',','').astype(int)
        df['Annual % Growth'] = df['Annual % Growth'].str.replace('%','').astype(float)
        
    else:
        print("None url conditions match. Problem.")
        return

    df = df[df["Year"] > "2009"]
    df = df[df["Year"] < "2019"]
    df = df.sort_values(by=['Year'], ascending=True)
    
    return df
```

Additionally, the scrapePages() function is defined which will take an input a list of URLs and then scrape each url and create a DataFrame of the desired contents for that page. <br>
Below the definition of this function, a list of all the URLs is created. This list is formatted for better readibility and will be used as input for the scrapePages function. <br>


```python
# function to iterate over given list of urls, scrape them and return dataframes
def scrapePages(list):
    for page in list:
        
        df = scrapper(page)
        formattedDF = formatter(df, page)

        print(formattedDF)
        
        dataframes.append(formattedDF)
```

This is repeated for all pages containing the desired data. Each DataFrame is then used to create a single DataFrame which contains all the macroeconomic indicators. <br>
The merging for each of those DataFrames is carried out by the mergeDFrames() function. <br>
By separating this entire process into smaller functions, this analysis achieves a good level of SOC which can prove useful for reusability and debugging in the future. <br> 
The DataFrame now contains only the required data in the correct form for further analysis.

```python
# to improve readability of pages in list below, defined a variable to hold the domain and the directory that doesnt change
base = "https://www.macrotrends.net/countries/GBR/united-kingdom/"

#append the page name to the above defined url
pages = [ base + "gdp-per-capita" ,
        base + "manufacturing-output" ,
        base + "inflation-rate-cpi" ,
        base + "gni-per-capita" ]

# list to hold dataframes of scraped pages
dataframes = []

#Grab dataframes and and merge them into one
def mergeDFrames(tables):
    # call scrapePages(pages) and pass the urls to scrappe. It will populate the dataframes list with dataframes to use below
    scrapePages(pages)

    #make the first dataframe the base which to merge others with
    combined = tables[0]
    
    i = 0
    
    # for each dataframe which isnt the first one, merge it with the base
    for frame in tables:
        if i > 0:
            combined = combined.merge(frame)
        i = i + 1
    return combined


df_macro = mergeDFrames(dataframes)

# save it locally
df_macro.to_csv('EconData_DataFrame.csv', encoding='utf-8')

df_macro
```

### 4.2 Empirical Analysis
#### 4.2.1 Correlation testing
The code below uses the .corr() method to calculate the correlation coefficient between the price of Bitcoin and each of the aforementioned macroeconomic data variables. <br>
<br>
Reading the results of each coefficient it can be deduced that: <br>
- GDP Per Capita USD does not exhibit a strong positive correlation whatsoever but neither does it entirely fall into the negative correlation value window. With a coefficient of -0.09, the relationship is close to 0 which is indecisive.
- Similarly, Manufacturing Output in Billions USD does not have a correlation. With a coefficient of -0.12, the relationship is close to 0 which is, again, indecisive.
- Inflation Rate % shows an interesting positive correlation of 0.09. Even though it does not pose a strong linear relationship, its difference from the other three macroeconomic variables shows that inflation may prove to have the most correlation with the price of Bitcoin out of them all.
- GNI Per Capita USD shows the lowest correlation coefficient of all macroeconomic variables at -0.3. This does close the gap between the coefficient and -1 which means that GNI per capita is most likely to NOT have a linear relationship with the price of Bitcoin.<br>
<br>
In statistics, the number -1.0 denotes a perfect negative correlation, while 0 denotes no connection and +1.0 denotes a perfect positive correlation. A complete negative correlation indicates that the relationship between two variables is always the total opposite (Nickolas, 2021). The results above only point towards the need of more in-depth analysis.

```python
#Get correlation between columns
for macro in colNamesMacro:
    stat = df_coin_macro[macro].corr(df_coin_macro['GBP per BTC'])
    print("Correlation between values in " , macro , " and GBP per BTC is " , stat)
```

#### 4.2.2 Pearson's r, Spearman's rho, Kendall's tau values testing
In the next series of tests, three coefficients are calculated to determine the relationship between each macroeconomic indicator the the price of Bitcoin. These are: <br>

- Pearson's r.<br> 
- Spearman’s rho.<br> 
- Kendall’s Tau.<br> 

After calculating each, they are followed by correlation matrix which is just another way of obtaining and representing the correlation between values and also uses Pearson's r coefficient (Bock, n.d.). <br> 

```python
# carry out hypothesis testing based on correlation and values of r, rho and tau
for macro in colNamesMacro:
    if macro != "GBP per BTC":
        x = df_coin_macro[macro]
        y = df_coin_macro['GBP per BTC']

        r = scipy.stats.pearsonr(x, y)    # Pearson's r
        rho  = scipy.stats.spearmanr(x, y)   # Spearman's rho
        tau = scipy.stats.kendalltau(x, y)  # Kendall's tau

        print('\033[1m' + "Tests of BTC against", macro, "result in the following values:" + '\033[0m')
        print("Pearson's r = " , r)
        print("Spearman's rho = " , rho)
        print("Kendall's tau = " , tau)
        # https://realpython.com/numpy-scipy-pandas-correlation-python/#example-scipy-correlation-calculation

        #corrcoef() returns the correlation matrix, which is a two-dimensional array with the correlation coefficients. 
        #The values on the main diagonal of the correlation matrix (upper left and lower right) are equal to 1. The upper left value corresponds to the correlation coefficient for x and x, while the lower right value is the correlation coefficient for y and y. They are always equal to 1.
        #However, what you usually need are the lower left and upper right values of the correlation matrix. These values are equal and both represent the Pearson correlation coefficient for x and y. In this case, it’s approximately 0.76.
        m = np.corrcoef(x, y)
        print("Correlation matrix is:")
        print(m)
        print()
        print()
```

The paragraphs below explain the range of the values and what they means in terms of relationships between variables. <br>
- The Pearson correlation coefficient r might be anything between +1 and -1. A value of 0 implies that the two variables have no relationship. A positive correlation is shown by a value larger than 0, whereas a negative association is indicated by a value less than 0 ("Pearson Product-Moment Correlation", n.d.).<br> 
- The correlation coefficients of Spearman's rho range from -1 to +1. A positive correlation of +1 implies that as one variable rises, the other tends to rise as well, and vice versa for negative correlations. Stronger links are represented by values near -1 or +1 than by values around zero (Frost, 2022).<br> 
- Kendall's Tau is a measure of data column relationships. The Tau correlation coefficient has a range of -1 to 1, with -1 denoting no relationship and 1 denoting a perfect one ("Kendall’s Tau - Simple Introduction", 2018). From the documentation it is visible the the default variant of the tau value is the Tau-b value, which is needed when analysing the data ("scipy.stats.kendalltau — SciPy v1.7.1 Manual", 2022). To determine if relationships are weak, moderate and strong, refer to (Mastrothanasis, 2020): <br> 
<br> Less than + or - 0.10: very weak
<br> + or -0.10 to 0.19: weak
<br> + or - 0.20 to 0.29: moderate
<br> + or - 0.30 or above: strong <br> 


Taking this into account, the following coefficients can be analysed:<br>
##### 4.2.2.1 BTC against GDP Per Capita USD
- Pearson's r (-0.09) Not significant enough value to show positive or negative relationship.<br>
- Spearman's rho (0.19) Not significant enough value to show positive or negative relationship, but slight tendency towards a positive relationship.<br>
- Kendall's tau (0.22) Moderate positive relationship.<br>


##### 4.2.2.2 BTC against Manufacturing Output in Billions USD
- Pearson's r (-0.12) Not significant enough value to show positive or negative relationship.<br>
- Spearman's rho (0.19) Not significant enough value to show positive or negative relationship, but slight tendency towards a positive relationship.<br>
- Kendall's tau (0.22) Moderate positive relationship.<br>


##### 4.2.2.3 BTC against Inflation Rate %
- Pearson's r (0.09) Not significant enough value to show positive or negative relationship.<br>
- Spearman's rho (-0.42) Negative value shows a somewhat negative relationship. <br>
- Kendall's tau (-0.30) Strong negative relationship.<br>


##### 4.2.2.4 BTC against GNI Per Capita USD

#### 4.2.3 Percentage change visual analysis
The final mean of analysis will employ a method to calculate the percentage changes in each variable before plotting them on graphs. This way, instead of comparing different metrics and values such as 10 and 30000, percentage changes should be more uniform and within a closer range from one another. Results should yield a clearer picture when plotted. To achieve this, the .pct_change() method is used to calculated all the values in columns based on the first row's value as a base. However, this leads to the first row cells converted into a Nan. Therefore, when plotting, the data will be taken from the second row onwards. By doing this change, the column names must be updated to reflect the changes of what values now represent.

```python
# for each column, calculated each new row's cell percentage change in value compared to previous
for col in colNames:
    #check if this has already ran. Otherwise many rows become NaN as a baseline for percentage comparison
    if col[0] != "NaN" :
        df_coin_macro[col] = df_coin_macro[col].pct_change()
        # in percent
        df_coin_macro[col] = df_coin_macro[col]*100
        
df_coin_macro = df_coin_macro.rename(columns={
    "Year":"Year","GDP Per Capita USD": "% Change in GDP Per Capita", 
    "Manufacturing Output in Billions USD": "% Change in Manufacturing Output", 
    "Inflation Rate %": "% Change in Inflation Rate",
    "GNI Per Capita USD" : "% Change in GNI Per Capita",
    "GBP per BTC" : "% Change in GBP per BTC"
})

df_coin_macro.to_csv('AllData_percentages_DataFrame.csv', encoding='utf-8')

df_coin_macro
```
